```python
from functools import reduce
# recursive functions
def display_range(lower,upper):
    """Outputs the numbers from lower to upper"""
    while lower <= upper:
        print(lower)
        lower += 1
```

```python
display_range(10,20)
```

```python
# we can achieve same behavior with recursion
def display_range2(lower,upper):
    if lower <= upper:
        print(lower)
        display_range2(lower + 1, upper)
```

```python
display_range2(10,20)
```

```python
% timeit display_range(10,20)
```

```python
%timeit display_range2(10,20)
```

```python
# second version is slightly slower
def our_sum(lower,upper):
    """returns the sum of numbers from lower to upper"""
    if lower > upper:
        return 0
    else:
        return lower + our_sum(lower + 1, upper)
```

```python
% timeit our_sum(10,20)
```

```python
%timeit sum(range(10,21)) # much faster
```

```python
def fact1(n):
    if n == 1:
        return n
    else:
        return n * fact1(n-1)
def fact2(n):
    return reduce(lambda a,b: a * b, range(1,n+1))
def fact3(n):
    result = n
    while n > 1:
        n -= 1
        result *= n
    return result
def fact4(n):
    def recurse(n,product):
        if n == 1: return product
        else: return recurse(n-1,n*product)
    recurse(n,1)
```

```python
%timeit fact1(20) 
%timeit fact2(20) 
%timeit fact3(20)# fact3 is the fastest
%timeit fact4(20)
```

```python
# classes.  more classes willb e covered later.  
class Counter:
    """Models a counter"""
    # class variable
    instances = []
    # constructor
    def __init__(self):
        Counter.instances.append(self)
        Counter.size = len(Counter.instances)
        self.reset()
    # mutator methods
    def reset(self):
        # sets the counter to 0
        self._value = 0
    def increment(self,amount = 1):
        """Adds amount to counter"""
        self._value += amount
    def decrement(self,amount = 1):
        """Subtracts amount from counter"""
        self._value -= amount
    def get_value(self):
        return self._value
    def __str__(self):
        """returns string representation of counter"""
        return str(self._value)
    def __eq__(self,other):
        """returns true if self equals other"""
        if self is other: return True
        if type(self) != type(other): return False
        return self._value == other._value
```

```python
# generators
def factors(n):
    """computes factors of a number"""
    result = []
    for k in range(1,n+1):
        if n % k == 0:
            result.append(k)
    return result
```

```python
factors(100)
```

```python
def factors2(n):
    for k in range(1,n+1):
        if n % k == 0:
            yield k
```

```python
list(factors2(100))
```

```python
for factor in factors2(100):
    print(factor)
```

```python
% timeit factors(100)
```

```python
% timeit factors2(100) # extremely fast.  
```

```python
%timeit list(factors2(100)) # slightly slower than above
```

```python
# much factor factor
def factors3(n):
    k = 1
    while k * k < n:
        if n % k == 0:
            yield k
            yield n // k # the quotient
        k += 1
    if k * k == n: # special case if n is a perfect square
        yield k
```

```python
%timeit factors3(100) # slightly faster than the previous case
```

```python
%timeit list(factors3(100)) # much faster but not sorted
```

```python
%timeit sorted(factors3(100)) # still much faster
```

```python
sorted(factors3(100))
```

```python
def fibonnaci(n):
    a = 0
    b = 1
    while n > 1:
        yield a # report value of a, during this pass
        future = a + b
        a = b # this will be next value reported
        b = future
        n -= 1
    
        
```

```python
list(fibonnaci(10))
```

```python
# alternative implementation of fibonnacci
def fib(n):
    a,b = 0,1
    while n > 1:
        yield a
        a, b = b, a + b
        n -= 1
```

```python
list(fib(10))
```
